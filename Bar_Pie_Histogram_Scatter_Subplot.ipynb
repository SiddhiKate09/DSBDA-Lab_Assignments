##Assignment comparing the mobile OS sale over the years

```python
years=['May','June','July','August','September','October','November','December']
Android=[1.44,1.51,1.61,1.64,1.64,1.58,1.43,1.46]
IOS=[17.8,17.7,18.34,18.41,18.12,18.49,18.43,17.4]
Windows=[27.97,27.01,26.03,25.53,26.67,26.83,25.67,25.73]
import matplotlib.pyplot as plt
plt.plot(years,Android,label="Android")
plt.plot(years,IOS,label="IOS")
plt.plot(years,Windows,label="Window")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
x=np.arange(len(years))
width=0.25
plt.bar(x-width,Android,width=width,label='Android')
plt.bar(x,IOS,width=width,label='IOS')
plt.bar(x+width,Windows,width=width,label='Window')
plt.xlabel('years')
plt.ylabel('Sale of various mobile os')
plt.xticks(x,years)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
android_total=sum(Android)
ios_total=sum(IOS)
window_total=sum(Windows)
labels=['Android','IOS','Windows']
values=[android_total,ios_total,window_total]
plt.title("Sales of Mobile Phone")
plt.pie(values,labels=labels,autopct='%1.1f%%')
plt.show()
```

```python
#using the scatter plot in data visualisation
labels=['Android','IOS','Windows']
colors=['blue','red','yellow']
plt.stackplot(years,Android,IOS,Windows,labels=labels,colors=colors)
plt.legend()
plt.title('Sales Distribution')
plt.xlabel('Mobiles operating System Sales')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
           
```
