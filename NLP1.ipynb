```python
# Morphology Add/Delete Table
# Sample words
words = ["run", "play", "happy"]
# Define morphological operations
data = [
    {"base": "run", "add": "-s", "result": "runs", "delete": "-s"},
    {"base": "play", "add": "-ing", "result": "playing", "delete": "-ing"},
    {"base": "happy", "add": "un- + -ness", "result": "unhappiness", "delete": "un- + -ness"}
]
# Print table
print("{:<15} {:<20} {:<20} {:<20}".format("Base Word", "Add", "Result Word", "Delete"))
print("-"*75)
for item in data:
    print("{:<15} {:<20} {:<20} {:<20}".format(
        item["base"],
        item["add"],
        item["result"],
        item["delete"]
    ))
```
